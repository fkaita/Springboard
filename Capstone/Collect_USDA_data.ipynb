```python
import requests
import json
import pandas as pd
```

```python
with open('token.json') as key_file:
    api_token = json.load(key_file)
```

```python
api_token
```

```python
headers = {"Authorization": "Bearer {}".format(api_token['key'])}

pub_id = 'j098zb08p' #id of oilcrop outlook

params = {'publicationId':'j098zb08p',
          'latest':'true'
         }

r = requests.get('https://usda.library.cornell.edu/api/v1/release/findByPubId/{}'.format(pub_id), headers=headers, params=params)

r.json()
```

```python
df = pd.read_excel(r.json()[0]['files'][0], sheet_name=4)
df
```
